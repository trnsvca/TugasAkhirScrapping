```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
link = "https://www.imdb.com/chart/bottom"
```

```python
page = requests.get(link)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_movies = soup.find_all('td', class_='titleColumn')
scraped_movies
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings
```

```python
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', "")
    ratings.append(rating)
ratings
```

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] = ratings
data.head()
```

```python
data.to_csv('IMDB Lowest Rated Movies.csv', index=False)
```
